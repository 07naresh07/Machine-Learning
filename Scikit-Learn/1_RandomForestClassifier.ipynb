```python
## Importing data 
import pandas as pd
import numpy as np
heart_disease = pd.read_csv('1_ML_Datas/heart-disease.csv')
heart_disease.head()
```

```python
#1. Getting data ready
x = heart_disease.drop('target', axis=1)
y = heart_disease['target']
```

```python
#2. Using RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
#Use default hyperparameters
clf.get_params()
```

```python
#3. Train_Test data 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
clf.fit(X_train, y_train)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
# Make the prediction
y_pred = clf.predict(X_test)
y_pred
```

```python
#4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
#Feature importance (Which factor is more important for decision making)
importances = clf.feature_importances_*100
feature_names = clf.feature_names_in_
for name, importance in zip(feature_names, importances):
    print(f'{name}: {importance:.2f}')
```

```python
#5. Improve a model
#Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 90, 10):
    print(f'Trying model with {i}_estimators...')
    clf = RandomForestClassifier(n_estimators=i)
    clf.fit(X_train, y_train)
    print(f'Model accuracy on test set: {clf.score(X_test, y_test)*100:.2f}%')
    print("")
```

```python
#6. Save a model and load it
import pickle
pickle.dump(clf, open('random_forest_model.pkl', 'wb'))
```

```python
import joblib
joblib.dump(clf, 'random_forest_model.joblib');
```
