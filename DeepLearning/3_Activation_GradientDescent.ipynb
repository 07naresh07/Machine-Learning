**Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Define the sigmoid activation function


```python
def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))
```

#### Define Sigmoid derivative, "the sigmoid gradient"


```python
def sigmoid_derivative(z):
    return sigmoid(z)*(1-sigmoid(z))
```

#### Define ReLU acctivation function


```python
def relu(z):
    return np.maximum(0, z)
```

```python
def relu_derivative(z):
    return np.where(z>0, 1, 0)
```

#### Define the number of input values
Generate 400 synthetic values between -10 and 10. The gradient would be calculated for each of these values.


```python
z = np.linspace(-10, 10, 400)
```

Next, let's compute the derivatives for Sigmoid and Relu activations and assign the values to 
**sigmoid_grad** and **relu_grad** respectively 


```python
sigmoid_grad = sigmoid_derivative(z)
relu_grad = relu_derivative(z)
```

Let's visualize the above data i.e. activation function and gradient.

```python
plt.figure(figsize=(10, 6))
# Sigmoid Plot
plt.subplot(1, 2, 1)
plt.plot(z, sigmoid(z), color='green', label='Sigmoid Activation')
plt.plot(z, sigmoid_grad, color='red', label='Sigmoid Derivative', ls='--')
plt.title('Sigmoid Activation and Gradient')
plt.xlabel('Input (z)')
plt.ylabel('Activation / Gradient')
plt.legend(loc='best')
plt.grid(True)

# Relu Plot
plt.subplot(1, 2, 2)
plt.plot(z, relu(z), color='green', label='ReLU Activation')
plt.plot(z, relu_grad, color='red', label='ReLU Derivative', ls='--')
plt.title('ReLU Activation and Gradient')
plt.xlabel('Input (z)')
plt.ylabel('Activation / Gradient')
plt.legend(loc='best')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Let's try with *hyperbolic tangent* activation**
* First Method

```python
def hyperbolic_tangent_first_method(z):
    return (np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z))

def hyperbolic_tangent_derivative_first_method(z):
    return 4/(np.exp(z)+np.exp(-z))**2

hyper_grad_first_method = hyperbolic_tangent_derivative_first_method(z)
```

* Second Method

```python
def hyperbolic_tangent_second_method(z):
    return np.tanh(z)

def hyperbolic_tangent_derivative_second_method(z):
    return 1-(np.tanh(z))**2

hyper_grad_second_method = hyperbolic_tangent_derivative_second_method(z)
```

Let's plot and compare the output of tanH function with the ReLU function for 100 synthetic values between -5 to 5

```python
z = np.linspace(-5, 5, 100)

# Gradient for Hyperbolic tangent activation function
hyper_grad = hyperbolic_tangent_derivative_second_method(z)

# Plot the curve
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(z, hyperbolic_tangent_second_method(z), color='g', label = 'Hyperbolic Activation')
plt.plot(z, hyper_grad, color='blue', label='Hyperbolic Derivative', ls='--')
plt.title('Hyperbolic Activation & Gradient Plot')
plt.xlabel('Input (z)')
plt.ylabel('Activation / Gradient')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(z, relu(z), color='red', label='ReLU Activation')
plt.plot(z, relu_derivative(z), color='blue', ls='--', label='Relu Derivative')
plt.title('ReLU Activation & Gradient Plot')
plt.xlabel('Input (z)')
plt.ylabel('Gradient')
plt.legend(loc='best')
plt.grid(True)

plt.tight_layout()
plt.show()
```
