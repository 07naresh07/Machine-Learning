```python
import pandas as pd
import numpy as np
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('CodesoftML/spam.csv', encoding='latin1')
```

```python
columns = [2, 3, 4]
df.drop(df.columns[columns], axis=1, inplace=True)
```

```python
df['target'] = df['v1'].map({'ham':0, 'spam':1}) 
```

```python
df.drop('v1', axis=1, inplace=True)
```

```python
df.rename(columns={'v2':'Emails'}, inplace=True)
```

```python
df.to_csv('Spam_Check.csv', index=False)
```

```python
df.head()
```

```python
df['Emails'] = df['Emails'].apply(lambda x: x.replace('\r\n', ''))
```

```python
df.Emails.iloc[2]
```

```python
df.info()
```

```python
stemmer = PorterStemmer()
corpus = []
stopwords = set(stopwords.words('english'))
```

```python
for i in range(len(df)):
    text = df['Emails'].iloc[i].lower()
    text = text.translate(str.maketrans('','',string.punctuation)).split()
    text = [stemmer.stem(word) for word in text if word not in stopwords]
    text = ' '.join(text)
    corpus.append(text)
```

```python
corpus[0]
```

```python
df.Emails[0]
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(corpus).toarray()
y = df['target']
```

```python
pd.DataFrame(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = RandomForestClassifier(n_jobs=-1)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

**Check a particular email to see if spam or not**

```python
email_to_check = df.Emails.values[700]
email_text = email_to_check.lower().translate(str.maketrans('','',string.punctuation)).split()
email_text = [stemmer.stem(word) for word in email_text if word not in stopwords]
email_text = ' '.join(email_text)
email_corpus = [email_text]
X_email = vectorizer.transform(email_corpus)
```
