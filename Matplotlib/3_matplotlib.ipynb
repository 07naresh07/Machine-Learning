## Two oprions for subplots

```python
#Subplot option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,8))
#Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10), c=np.random.randint(0, 100, 10), cmap='viridis', alpha=0.5, sizes=np.random.randint(100, 1000, 10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values(), width=0.4)
ax4.hist(np.random.randn(500), color='hotpink')
plt.show()
```

```python
#Subplot option 2:
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,8))
x=np.linspace(0,100,10)
ax[0,0].plot(x, x**3)
ax[0,1].scatter(x, np.log(x), c=np.random.randint(20, 100, 10), cmap='viridis', sizes = 5*np.random.randint(10, 200, 10), alpha=0.4)
ax[1,0].barh(nut_butter_prices.keys(), nut_butter_prices.values(), height=0.4, color='red')
ax[1,1].hist(np.random.randint(-100, 100, 1000))

plt.show()
```

## Which method to use to plot pandas DataFrame for histogram (Pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
#Pyplot method (Direct method)
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
            figsize=(10,6))
plt.show()
```

```python
# OO method mixed with pyplot (direct) method
fig, axi = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=axi)
axi.set_xlim([45,100])  ##We can use tuples or keywords like left=45, right=100
# axi.set_xlim(left=45, right=120)
plt.show()
```

```python
## OO method from scratch (OO: Object Oriented Method)
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
#Customize data
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')
#Add legend
avg_chol = over_50['chol'].mean()
ax.legend(*scatter.legend_elements(), title='Target')

#Add horizontal line for average cholesterol
ax.axhline(avg_chol, color='red', ls='--', lw=2, label=f'Average Cholesterol {avg_chol: .3f}')
ax.legend(loc='upper left')

plt.show()
```

```python
over_50.head()
```

```python
fig, axi = plt.subplots(figsize=(10,6))
scatter_plot = axi.scatter(x=over_50['age'],
                           y=over_50['thalach'],
                           c=over_50['thal'])
#Adding titles and axis names
axi.set(title='Age vs Heart Rate Plot',
        xlabel='Age',
        ylabel='Heart Rate')
#Adding legend
axi.legend(*scatter_plot.legend_elements(), title='Leged')

avg_heartRate = over_50['thalach'].mean()
#Adding average horizontal line at the center
axi.axhline(avg_heartRate, ls='-.', lw=2, color='hotpink', label=f'Average Heart Rate {avg_heartRate: .2f}')
axi.legend(loc='lower right')

plt.show()
```

```python
#Subplot of age, chol, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)
#Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='plasma')
#Customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')
#Add legend
ax0.legend(*scatter.legend_elements(), title='Target')
#Add mean line
ax0.axhline(over_50['chol'].mean(), ls='dashed', color='red', lw=2)

# Add Data to ax1
scatter = ax1.scatter(x=over_50['age'],
            y=over_50['thalach'],
            c=over_50['target'],
            cmap='inferno')

#Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

#Add a legend
ax1.legend(*scatter.legend_elements(), title='Target')

#Add mean line
ax1.axhline(over_50['thalach'].mean(), ls='-.', color='red', lw=2)

#Add a title to the figure
fig.suptitle('Heart Diseases Analysis', font='serif', fontsize=15, style='italic', weight='bold', color='blue')
plt.show()
            
```
