```python
import pandas as pd
import numpy as np

sampleData = 1500
np.random.seed(42)

data = {
    'CustomerID': [f"C{str(i).zfill(4)}" for i in range(1, sampleData+1)],
    'Age': np.random.randint(20, 60, sampleData),
    'Gender': np.random.choice(['Male', 'Female'], sampleData),
    'MembershipType': np.random.choice(['Bronze', 'Silver', 'Gold'], sampleData, p=[0.5, 0.3, 0.2]),
    'LastPurchaseAmount': np.random.randint(10, 2000, sampleData),
    'AvgResponseTime': np.random.uniform(0.5, 5.0, sampleData).round(1),
    'NumSupportTickets': np.random.randint(0, 10, sampleData),
    'Satisfaction': np.random.choice(['Satisfied', 'Unsatisfied'], sampleData, p=[0.6, 0.4])
}
```

```python
df = pd.DataFrame(data)
```
