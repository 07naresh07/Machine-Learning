```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_matrix(n, threshold=0.01, epsilon=0.01):
    A = np.random.uniform(0,10, (n,n))
    A = np.sqrt(A)*np.exp(np.sin(A))
    determinant = np.linalg.det(A)
    while determinant<=0:
        A+=epsilon*np.eye(n)
        determinant = np.linalg.det(A)
    eigenValue, eigenVector = np.linalg.eig(A)
    return A, determinant, eigenValue, eigenVector
```

```python
def plot_matrix(eigenvalues):
    plt.figure(figsize=(10,6))
    plt.scatter(eigenvalues.real, eigenvalues.imag, label='Eigen Values', color='blue')
    plt.axhline(0, lw=0.5, ls='--', color='red')
    plt.axvline(0, lw=0.5, ls='--', color='red')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginery Part')
    plt.title('Eigen Values on the complex plane')
    plt.legend()
    plt.show()
```

```python
if __name__=="__main__":
    n=5
    A, determinant, eigenvalues, eigenvectors = generate_matrix(n)
    print("Modified Matrix A:")
    print(A)
    print("\nDeterminant of A:", determinant)
    print("\nEigenvalues of A:", eigenvalues)
    print("\nEigenvectors of A:")
    print(eigenvectors)
    plot_matrix(eigenvalues)
```
