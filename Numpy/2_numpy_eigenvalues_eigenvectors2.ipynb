```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
def generate_matrix(n, epsilon=0.01):
    B = np.random.uniform(0, 10, (n,n))
    A = B + B.T
    while np.any(np.linalg.eigvals(A))<0:
        A+= epsilon*np.eye(n)
    eigenvalues, eigenvectors = np.linalg.eig(A)
    condition_number = np.max(eigenvalues)/np.min(eigenvalues)
    return A, eigenvalues, eigenvectors, condition_number
```

```python
def matrix_plot(eigenvalues):
    plt.figure(figsize=(10,6))
    plt.plot(eigenvalues, np.zeros_like(eigenvalues), marker='o', color='green', label='Eigenvalues')
    plt.axvline(np.min(eigenvalues), ls='--', color='red', label='Smallest Eigenvalues')
    plt.axvline(np.max(eigenvalues), ls='--', color='blue', label='Largest Eigenvalues')
    plt.xlabel('Eigenvalue')
    plt.ylabel('Magnitude')
    plt.title('Eigenvalues of the Matrix')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
if __name__== "__main__":
    n=5
    A, eigenvalues, eigenvectors, condition_number = generate_matrix(n)
    print("Symmetric Positive Definite Matrix A:")
    print(A)
    print("\nEigenvalues of A:", eigenvalues)
    print("\nCondition Number of A:", condition_number)
    matrix_plot(eigenvalues)
```
